 ## **Problem Statment**: 

With many new home buyers, many times it can be very confusing experience. Especially if they do not know what kind of house their budget could potentially buy. For many, certain feature of home are more important than others. As a example, someone may want a pool in the house more than they want a 3 car garage. Adding and eliminating features changes the price of housing and therefore will determine if a buyer could potentially buy the home.Therefore will look at Ames Housing Data and select features & model regressions to produce the most accurate model to predict the ‘Sale Price’ of a house based on different features of the house. In this way we can help a new home buyer select features by priority to find them a house that they can afford while still letting them have the house they always wanted. 

We will be looking about 81 features:
1. Qualitative: Stuff like Garage Area, 2 or 3 car Garages, Pool Quality ect.
2. Quantitative:  Garage Area, Total Basement surface area ect.. 

These features will let us know 
1. Given some list of features a client is looking for in a house, what kind of housing price could they expect?
2. What model would be best to use?

````python
def Remove_Outlier_Indices(df): # Got from https://stackoverflow.com/questions/46245035/pandas-dataframe-remove-outliers
    Q1 = df.quantile(0.30)
    Q3 = df.quantile(0.70)
    IQR = Q3 - Q1
    trueList = ~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR)))
    return trueList
````

**Preprocessing and Modeling 1**

````python
train_index = 1587
````

````python
df = pd.DataFrame(data = {x:train_test_df[:train_index][x].nunique() for x in list(train_test_df.columns)}, index = [0])
````

````python
columns_to_dummify = list(train_test_df[:train_index][list(df.T[df.T[0] < 100].T.columns)].select_dtypes(include=['object']).columns)
columns_to_dummify
````

````python
train_test_df_dummy = pd.get_dummies(train_test_df, columns =columns_to_dummify, prefix = columns_to_dummify  ,drop_first = True)
````

````python
X_train, X_test, y_train, y_test = train_test_split(train_test_df_dummy[:train_index].drop(columns = ['Sale Price']), train_test_df_dummy[:train_index]['Sale Price'], test_size=0.2, random_state=42)
````

````python
lr = LinearRegression()
````

````python
lr.fit(X_train, y_train)
````

````python
y_pred = lr.predict(X_test)
````

````python
mse = mean_squared_error(y_test,y_pred)
mse**(1/2)
````

````python
r2_score(y_test,y_pred)
````

````python
sns.distplot(y_test-y_pred);
plt.title("Distribution of Residuals")
````

````python
print('The root mean squared error (RMSE) is: ' + str(mse**(1/2)) +', which means our model is 89% accurate.\n')
print('The R^2 score is '+ str(r2_score(y_test,y_pred)) +', which means our model explains 89% of the variation of the data/fits 89% of the data ')
````

````python
lasso_reg = Lasso()
````

````python
lasso_reg.fit(X_train, y_train)
````

````python
L_y_pred = lasso_reg.predict(X_test)
````

````python
lmse =(mean_squared_error(y_test,L_y_pred))**(0.5)
lmse
````

````python
r2_score(y_test,L_y_pred)
````

````python
sns.distplot(y_test-L_y_pred);
plt.title("Distribution of Residuals")
````

````python
print('The root mean squared error (RMSE) is: ' + str(lmse) +', which means our model is 81% accurate.\n')
print('The R^2 score is '+ str(r2_score(y_test,L_y_pred)) +', which means our model explains 67% of the variation of the data/fits 67% of the data ')
````
